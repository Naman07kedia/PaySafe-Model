```python
import pandas as pd

# Step 1: Load raw dataset
df = pd.read_csv(r"C:\Users\naman\PaySafe UPI Fraud Detection\Data\Raw_Data.csv")
# Using r prefix to treat the string as a raw string, or use double backslashes or forward slashes

# Step 2: Clean data
df.drop_duplicates(inplace=True)
df.fillna(method="ffill", inplace=True)
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Step 3: Save cleaned dataset
df.to_csv(r"C:\Users\naman\PaySafe UPI Fraud Detection\Data\cleaned_Data.csv", index=False)
# Using r prefix here as well

print("Cleaned dataset saved at C:\\Users\\naman\\PaySafe UPI Fraud Detection\\Data\\cleaned_Data.csv")
# In print statements, use double backslashes to escape the backslash character
```
